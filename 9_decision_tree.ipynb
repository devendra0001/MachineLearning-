<h2 style='color:blue' align="center">Decision Tree Classification</h2>

```python
import pandas as pd
```

```python
df = pd.read_csv("salaries.csv")
df
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
```

```python
target = df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company']= le_company.fit_transform(inputs['company'])
inputs['job']= le_job.fit_transform(inputs['job'])
inputs['degree']=le_degree.fit_transform(inputs['degree'])
```

```python
labels_company=dict(zip(le_company.classes_,range(len(le_company.classes_))))
labels_job=dict(zip(le_job.classes_,range(len(le_job.classes_))))
print(labels_company,labels_job)
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(inputs,target)
```

**Is salary of Google, Computer Engineer, Bachelors degree > 100 k ?**

```python
model.predict([[2,1,0]])
```

**Is salary of Google, Computer Engineer, Masters degree > 100 k ?**

```python
model.predict([[2,1,1]])
```
