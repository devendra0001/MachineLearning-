```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Load Dataset
data = pd.read_csv("iot_device_data.csv")
```

```python
# Quick look at the data
print(data.head(30))
```

```python
# Define the features and the target
# Assuming the target is 'status' and it's categorical (e.g., 'normal' or 'faulty')
X = data[['temperature', 'humidity', 'pressure', 'vibration']]  # Use relevant IoT features
y = data['status']
```

```python
# Convert categorical target to numerical (if necessary)
y = pd.factorize(y)[0]
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize the Random Forest Classifier
model = RandomForestClassifier(n_estimators=11, random_state=42)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
y_pred
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
